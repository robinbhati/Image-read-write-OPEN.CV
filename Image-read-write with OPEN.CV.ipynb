```python
import cv2

# Load the image

img = cv2.imread("lena.jpg", 1)

# We have 3 ways to read image with 1= colorful img, with 0=grayscale, with -1= will load image with alpha channels 


#img = cv2.imread("lena.jpg", 1)  :with 1= colorful img
#img = cv2.imread("lena.jpg", -1): with -1= will load image with alpha channels

#Below are the matrix of image 
```

```python


# Check if the image was loaded successfully
if img is not None:
    # Display the image in a window
    cv2.imshow("Image Window", img)
    
    # Wait for a key event and store the key pressed it is used basically to extend the time of window frame
    key = cv2.waitKey(0)
    
    # Check if the 'q' key was pressed 
    if key == 113:
        cv2.destroyAllWindows()  # Close the window if 'q' is pressed
else:
    print("Image not loaded successfully")
```

```python
# Display the image in the notebook cell
display(Image(data=cv2.imencode('lena.jpg', img)[1].tobytes()))
```
